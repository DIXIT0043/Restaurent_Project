```python
import pandas as pd
import numpy as np
```

```python
df2 =pd.read_csv(r'Dataset.csv')
df2
```

```python
# already checked the missing values and duplicate values in Task 1
```

```python
import re
def remove_special_chars(text):
    if isinstance(text, str):
        return re.sub(r'[^\w\s]', '', text)
    else:
        return text
```

```python
df = df2.applymap(remove_special_chars)
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import folium
```

```python
m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=12)
```

```python
for i, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], 
                  popup=f"{row['Restaurant Name']} - {row['Locality']}",
                  icon=folium.Icon(color='blue')).add_to(m)
```

```python
m.save('restaurant_map.html')
```

```python
grouped_by_city = df.groupby('City').size().reset_index(name='Number of Restaurants')
```

```python
plt.figure(figsize=(12, 6))
plt.bar(grouped_by_city['City'], grouped_by_city['Number of Restaurants'], color='skyblue')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.title('Number of Restaurants by City')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
avg_ratings_by_city = df.groupby('City')['Aggregate rating'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(avg_ratings_by_city['City'], avg_ratings_by_city['Aggregate rating'], color='green')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title('Average Rating by City')
plt.xticks(rotation=90)
plt.tight_layout()
```
