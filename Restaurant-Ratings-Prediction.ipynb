```python
import re
def remove_special_chars(text):
    if isinstance(text, str):
        return re.sub(r'[^\w\s]', '', text)
    else:
        return text
```

```python
features = ['Country Code', 'City', 'Cuisines', 'Average Cost for two', 'Has Table booking', 'Has Online delivery', 'Price range', 'Votes']
```

```python
X = df[features]
y = df['Aggregate rating']
```

```python
categorical_features = ['City', 'Cuisines', 'Has Table booking', 'Has Online delivery']
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot = OneHotEncoder()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer([("one_hot", one_hot, categorical_features)], remainder="passthrough")
```

```python
X_transformed = transformer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
rmse = mean_squared_error(y_test, y_pred, squared=False)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared: {r_squared}")
```

```python
coefficients = model.coef_
```

```python
feature_names = transformer.named_transformers_['one_hot'].get_feature_names_out()
```

```python
all_feature_names = feature_names.tolist() + ['Country Code', 'Average Cost for two', 'Price range', 'Votes']
```

```python
coefficients_df = pd.DataFrame({
    'Feature': all_feature_names,
    'Coefficient': coefficients
})
```

```python
coefficients_df['Absolute Coefficient'] = coefficients_df['Coefficient'].abs()
coefficients_df = coefficients_df.sort_values(by='Absolute Coefficient', ascending=False)
coefficients_df
```
